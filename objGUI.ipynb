```python
import numpy as np
import cv2
from tkinter import *
import tkinter
from tkinter import filedialog

base = tkinter.Tk()
base.geometry('500x400')



def process(cap):
    
    global net, classes, layer_names, output_layers, colors, saveVid,lmain
    saveVidFlag = saveVid.get() 
    
    if saveVidFlag:
        ret,frame = cap.read()
        frame = cv2.resize(frame, (0, 0), fx=1, fy=1)

        size = (frame.shape[1],frame.shape[0])

        result = cv2.VideoWriter('processed.avi', cv2.VideoWriter_fourcc(*'MJPG'), 15, size)

    flag = True
    while cap.isOpened():
        print("saveVid--> ",saveVidFlag)
        if flag:
            ret, frame = cap.read()
            if not ret:
                break
            frame=cv2.resize(frame,(0,0),fx=1,fy=1)
            
            height, width, channels = frame.shape

            # Detecting objects
            blob = cv2.dnn.blobFromImage(frame,scalefactor= 0.00392, size=(320,320), mean=(0, 0, 0), swapRB=True, crop=False)#0.00392

            net.setInput(blob)

            outs = net.forward(output_layers)
            #print(len(outs))
            # Showing informations on the screen
            class_ids = []
            confidences = []
            boxes = []

            for out in outs:
                for detection in out:
                    
                    scores = detection[5:]
                    class_id = np.argmax(scores)
                    confidence = scores[class_id]

                    if confidence > 0.5:
                        # Object detected
                        center_x = int(detection[0] * width)
                        center_y = int(detection[1] * height)
                        w = int(detection[2] * width)
                        h = int(detection[3] * height)

                        # Rectangle coordinates
                        x = int(center_x - w / 2)
                        y = int(center_y - h / 2)

                        boxes.append([x, y, w, h])
                        confidences.append(float(confidence))
                        class_ids.append(class_id)

            indexes = np.squeeze(cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4))


            
            font = cv2.FONT_HERSHEY_PLAIN
            #print(len(indexes))
            for i in range(len(boxes)):
                if i in indexes:
                    x, y, w, h = boxes[i]
                    label = str(classes[class_ids[i]])
                    color = colors[i]
                    cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
                    cv2.putText(frame, label, (x, y + 10), font, 0.8, color, 2)
            if saveVidFlag:
                result.write(frame)
            

            cv2.imshow("Yolov3 Object Detection", frame)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        flag = not flag
    if saveVidFlag:
        result.release()
        l = Label(base, text = "Video Saved at:  C:\\Users\\khede\\OneDrive\\Desktop\\processed.avi") 
        l.config(font =("Courier", 8))
        l.place(x=50,y=240)
    cap.release()
    cv2.destroyAllWindows()

def camera():
    cap = cv2.VideoCapture(0)
    if cap.isOpened() == False:
        print("Error opening video  file")
    else:
        process(cap)
    
    
def video(pth):    
    
    cap = cv2.VideoCapture(pth)
    

    if cap.isOpened() == False:
        print("Error opening video  file")
    else:
        process(cap)
    
    
# Load Yolo
net = cv2.dnn.readNet("yolov3.weights", "yolov3.cfg")
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_OPENCV)
net.setPreferableTarget(cv2.dnn.DNN_TARGET_CPU)
classes = []

with open("coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]

layer_names = net.getLayerNames()
#print(net.getUnconnectedOutLayers())

output_layers = [layer_names[i - 1] for i in np.squeeze(net.getUnconnectedOutLayers())]


colors = np.random.uniform(0, 255, size=(len(classes), 3))
print(len(layer_names))


t = Label(base, text = "Object Recognition Using openCV") 
t.config(font =("Courier", 18))
t.place(x=60,y=20)

x = Button(base, text ='Camera', height = 3, bg="#61ffc8", width = 20,command = lambda:camera())
x.place(x=170,y=60)
y = Button(base, text ='Select a file', height = 3, bg="#61ffc8",width = 20,command = lambda:video(filedialog.askopenfile(initialdir="/").name))
y.place(x=170,y=130)
saveVid = IntVar()
C1 = Checkbutton(base, text="Save?", variable=saveVid, onvalue = 1,offvalue = 0,height = 2, width = 30)
C1.place(x=120,y=190)


base.mainloop()


```
